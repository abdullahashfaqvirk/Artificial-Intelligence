```python
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers=headers)
data = StringIO(req.text)
```

```python
pd.read_csv(data).head()
```

### 03. Sep Parameter

```python
# Using movie titles metadata 

pd.read_csv(data, sep='\t', names=['sno','name','release_year','rating','votes','genres']).head()
```

### 04. Index Col Parameter

```python
# Set the 'index_col' parameter to 'column_name' to use it as the index column of the DataFrame

pd.read_csv('file_name.csv', index_col='column_name')
```

### 05. Header Parameter

```python
# Using test dataset

pd.read_csv(data, header=1).head()
```

### 06. Use Cols Parameter

```python
# Using aug_train dataset

pd.read_csv(data, usecols=["enrollee_id", "gender", "education_level"]).head()
```

### 07. Squeeze Parameters

```python
# Convert pandas dataframe to pandas series
# Using aug_train dataset

pd.read_csv(data, usecols=['gender'], squeeze=True).head(10)
```

### 08. Skiprows/nrows Parameter

```python
# Using aug_train dataset

pd.read_csv(data, skiprows=[2, 3]).head()
```

```python
pd.read_csv(data, nrows=100)
```

### 09. Encoding Parameter

```python
# Using zomato dataset

pd.read_csv(data, encoding='latin-1').head()
```

### 10. Skip Error Bad Lines

```python
# Set the 'error_bad_lines' parameter to False, to skip lines with bad data instead of raising an error.
# Skip the rows which are more then cols

pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", error_bad_lines=False)
```

### 11. dtypes Parameter

```python
# Converting one dtype to another dtype
# Using aug_train dataset

# The target column is in float (values 0 and 1), which must be converted to int to save space.
# pd.read_csv(data).info()

pd.read_csv(data, dtype={'target': int}).info()
```

### 12. Handling Dates

```python
# Using ipl matches dataset
# In this dataset date coulumn is an object(string) which must be converted into date datatype

pd.read_csv(data).info()
```

```python
# Also, if there is a need to merge or combine multiple date columns into a single datetime column, 
# the parse_dates parameter can be helpful.

pd.read_csv(data, parse_dates=['date']).info()
```

### 13. Convertors

```python
def rename_vals(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
# Using ipl matches dataset
# convertors = {'column name': function name}

pd.read_csv(data, converters={'team1': rename_vals}).head()
```

### 14. na_values Parameter

```python
# Using aug_train dataset
# use for converting '-' or placeholders to convert NaN value
# na_values=['Male',] -> converting Male to NaN value

pd.read_csv(data, na_values=['Male',])
```

### 15. Loading a Huge Dataset in Chunks

```python
# Using aug_train dataset

dfs = pd.read_csv(data, chunksize=5000)

for chunks in dfs:
    print(chunks.shape)
```
